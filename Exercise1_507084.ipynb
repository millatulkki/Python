```python
## Task 3
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Load the data matrix
data2 = np.genfromtxt("task_III_data.txt",delimiter=",")

# first column = x, second = y
# squeezed to make 1D arrays
x = data2[:,[0]]
y = data2[:,[1]]
x_1D = x.squeeze()
y_1D = y.squeeze()

# calculate linear and cupic interpolation functions with interp1d
lin = interp1d(x_1D,y_1D,kind="linear")
cub = interp1d(x_1D,y_1D,kind="cubic")

# Plot the loaded data and both interpolated functions
y_linear = lin(x)
y_cubic = cub(x)
plt.plot(x,y,'o',x,y_linear, '-',x,y_cubic, '--')
plt.title("Input data and interpolated functions",fontsize=20)
plt.xlabel("Input value")
plt.ylabel("Function value")
plt.legend(['data', 'linear', 'cubic'], loc='best')
plt.show()
```
