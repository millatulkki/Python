```python
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the data matrix from comma-delimited .txt-file
data2 = np.genfromtxt("task_III_data.txt",delimiter=",")

# In the data first column = x, second = y
# Squeezed to make 1D arrays
x = data2[:,[0]]
y = data2[:,[1]]
x_1D = x.squeeze()
y_1D = y.squeeze()

# Calculating linear and cupic interpolation functions
lin = interp1d(x_1D,y_1D,kind="linear")
cub = interp1d(x_1D,y_1D,kind="cubic")
# new x's for plotting the functions
xnew = np.linspace(0,40,num=41,endpoint=True)

# Plot the loaded data and both interpolated functions
y_linear = lin(xnew)
y_cubic = cub(xnew)
plt.plot(x,y,'o',xnew,y_linear, '-',xnew,y_cubic, '--')
plt.title("Input data and interpolated functions",fontsize=20)
plt.xlabel("Input value")
plt.ylabel("Function value")
plt.legend(['data', 'linear', 'cubic'], loc='best')
plt.show()
```
